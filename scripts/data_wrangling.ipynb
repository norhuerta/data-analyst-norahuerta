```python
import pandas as pd

# Load the raw dataset from your call center data
raw_data = pd.read_csv('data/call_center_raw.csv')

# Step 1: Handle missing values
cleaned_data = raw_data.fillna(0)

# Step 2: Remove duplicates
cleaned_data = cleaned_data.drop_duplicates()

# Step 3: Standardize the 'date' column
cleaned_data['date'] = pd.to_datetime(cleaned_data['date'])

# Save cleaned data
cleaned_data.to_csv('data/cleaned_call_center_data.csv', index=False)
```
